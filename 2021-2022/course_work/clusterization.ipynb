```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
X = pd.read_csv("customer_clast.csv", sep = ',', skipfooter = 1975)
```

```python
X
```

```python
X = X.drop(["ID"], axis = 1)
X
```

```python
int(X["Age"].mean())
```

```python
X["Income"].mean()
```

```python
for columns in X:
    plt.figure(figsize = (8, 4))
    sns.countplot(data = X, x = columns)
```

```python
X.corr()
```

```python
sns.heatmap(X.corr(), cmap = "RdBu", annot = True);
```

```python
sns.pairplot(X)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dataset_std = scaler.fit_transform(X)
```

```python
dataset_std
```

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(dataset_std,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(dataset_std)
```

```python
prediction=kmeans.fit_predict(dataset_std)
prediction
```

```python
new_data=X.copy()
new_data['cluster']=prediction
new_data
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(new_data['Education'],new_data["Age"],c=new_data['cluster'],cmap='rainbow', s = 100)
plt.xlabel('Education')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(new_data['Occupation'],new_data["Income"],c=new_data['cluster'],cmap='rainbow', s = 100)
plt.xlabel('Occupation')
plt.ylabel('Income')
plt.show()
```
