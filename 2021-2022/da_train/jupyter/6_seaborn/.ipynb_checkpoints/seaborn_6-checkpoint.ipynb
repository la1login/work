## Adding a third variable with hue

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,
               hue='smoker')
plt.show()
```

### Hue order

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,
               hue='smoker',
               hue_order=['No', 'Yes'])
plt.show()
```

### Color setting

```python
hue_colors = {'Yes': 'black', 'No': 'red'}
sns.scatterplot(x='total_bill', y='tip', data=tips,
               hue='smoker',
               hue_order=['No', 'Yes'],
               palette=hue_colors)
plt.show()
```

```python
hue_colors = {'Yes': '#808080', 'No': '#00FF00'}
sns.scatterplot(x='total_bill', y='tip', data=tips,
               hue='smoker',
               hue_order=['No', 'Yes'],
               palette=hue_colors)
plt.show()
```

### Hue with count plots

```python
sns.countplot(x='smoker', data=tips, hue='sex')
plt.show()
```

# Introduction ro relational plots and subplots

### Two types of  relational plots: scatter and line

```python
sns.relplot(x='total_bill', y='tip', 
            data=tips, kind='scatter')
plt.show()
```

### Making subplots

```python
sns.relplot(x='total_bill', y='tip', 
            data=tips, kind='scatter',
           col='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', 
            data=tips, kind='scatter',
           row='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', 
            data=tips, kind='scatter',
           col='smoker', row='time')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', 
            data=tips, kind='scatter',
           col='day', col_wrap=2)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', 
            data=tips, kind='scatter',
           col='day', col_wrap=2,
           col_order=['Thur',
                     'Sat',
                     'Fri',
                     'Sun'])
plt.show()
```

## Customizing scatter plots

### similar kwargs for scatterplot and relplot

### Point size and color

```python
# size is variable from df
sns.relplot(x='total_bill', y='tip', data=tips,
               kind='scatter', size='size', hue='size')
plt.show()
```

### Point style

```python
sns.relplot(x='total_bill', y='tip', data=tips,
           kind='scatter', hue='smoker', style='smoker')
plt.show()
```

### Point transparency

```python
sns.relplot(x='total_bill', y='tip', data=tips,
           kind='scatter', alpha=0.4)
plt.show()
```

## Introduction to line plots

```python
head = ['hour', 'location', 'no_2_mean', 'station']
location = ['east', 'north', 'south', 'west']
data = []
for i in range(24):
    for j in range(4):
        data.append([i + 1, location[j], random.uniform(13, 35), random.randint(28079001, 28079100)])
df = pd.DataFrame(data, columns=head)
df
```

```python
sns.relplot(x='hour', y='no_2_mean', data=df,
           kind='line')
plt.show()
```

### Subgroups

```python
sns.relplot(x='hour', y='no_2_mean', data=df,
           kind='line', style='location',
           hue='location', markers=True)
plt.show()
```

### Multiple obseravations per x-value

### Все значения уровня оксида азота объединены и жирной линией показано среднее значение, область - доверительный интервал

```python
sns.relplot(x='hour', y='no_2_mean', data=df, kind='line')
plt.show()
```

### Заменим доверительный интервал на стандартное отклонение

```python
sns.relplot(x='hour', y='no_2_mean', data=df, kind='line',
           ci='sd')
plt.show()
```

## Box plots

```python
sns.catplot(x='time', y='total_bill', data=tips, kind='box')
plt.show()
```

### Change the order

```python
sns.catplot(x='time', y='total_bill', data=tips, kind='box',
           order=['Dinner', 'Lunch'])
plt.show()
```

### Исключение выбросов

```python
sns.catplot(x='time', y='total_bill', data=tips, kind='box',
           sym='')
plt.show()
```

### Changing whis (линии квартилей)
default 1.5 * IQR

whis=2.0 means 2 * IQR

whis=[5, 95]

whis=[0, 100] - max values

```python
sns.catplot(x='time', y='total_bill', data=tips, kind='box',
           whis=[0, 100])
plt.show()
```

### Нет выбросов тк все значения включены

## Point plots

### Show mean of quantitative variable

### Lines from points show 95% confidence interval

### x-axis usually categorical

```python
df_cat
```

```python
df_cat['age_group'] = df_cat['Age'] % 3
```

```python
sns.catplot(x='age_group', y='Annual Income (k$)', data=df_cat,
           hue='Gender', kind='point')
plt.show()
```

### Disconnecting the points

```python
sns.catplot(x='age_group', y='Annual Income (k$)', data=df_cat,
           hue='Gender', kind='point', join=False)
plt.show()
```

### Displaying the median isntead of mean

```python
# default
sns.catplot(x='smoker', y='total_bill', data=tips, 
            kind='point')
plt.show()
```

```python
# change to median
from numpy import median

sns.catplot(x='smoker', y='total_bill', data=tips,
           kind='point', estimator=median, col='day',col_wrap=2)
plt.show()
```

### Customizing the confidence intervals

```python
sns.catplot(x='smoker', y='total_bill', data=tips,
           kind='point', capsize=0.2)
plt.show()
```

### Delete confidence intervals

```python
sns.catplot(x='smoker', y='total_bill', data=tips,
           kind='point', ci=None)
plt.show()
```

# Customizing seaborn plots

## Changing plot style and color

```python
sns.set_style('whitegrid')
sns.catplot(x='smoker', y='total_bill', data=tips,
           kind='point')
plt.show()
```

```python
sns.set_style('dark')
sns.catplot(x='smoker', y='total_bill', data=tips,
           kind='point')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.catplot(x='smoker', y='total_bill', data=tips,
           kind='point')
plt.show()
```

### Changing the palette

```python
sns.set_style('white')
sns.set_palette('RdBu')
order = [1, 3, 5, 2, 4]
sns.catplot(x='happiness', data=df_cat, kind='count',
           order=order)
plt.show()
```

### Custom palettes

```python
sns.set_palette(palette)
```

```python
sns.catplot(x='happiness', data=df_cat, kind='count',
           order=order)
plt.show()
```

### Changing the scale

```python
sns.set_context('talk')
sns.catplot(x='happiness', data=df_cat, kind='count',
           order=order)
plt.show()
```

## Adding titles and labels 1

### Add title to FacetGrid object

```python
sns.set_context('notebook')
g = sns.catplot(x='smoker', y='total_bill', data=tips,
               kind='box')
g.fig.suptitle('Mean total bill for smokers and no smokers',
              y=1.03)
plt.show()
```

## Adding titles and labels 2

### Adding title to AxesSubplot object

```python
g = sns.boxplot(x='smoker', y='total_bill', data=tips)
g.set_title('Mean total bill for smokers and no smokers', y=1.04)
plt.show()
```

### Adding titles to Subplots

```python
g = sns.catplot(x='smoker', y='total_bill', data=tips,
           kind='box', col='time')
g.fig.suptitle('Mean total bill for smokers and no smokers\n by day',
              y=1.05) # main title for all plots
g.set_titles('This is {col_name}')
plt.show()
```

### Adding axis labels

```python
g = sns.catplot(x='smoker', y='total_bill', data=tips,
               kind='box')
g.set(xlabel='Smokers',
     ylabel='Bill')
plt.show()
```

### Rotating x=axis tick labels

```python
g = sns.catplot(x='smoker', y='total_bill', data=tips,
                kind='box')
plt.xticks(rotation=45)
plt.show()
```
