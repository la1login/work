## Introduction to line plots

```python
head = ['hour', 'location', 'no_2_mean', 'station']
location = ['east', 'north', 'south', 'west']
data = []
for i in range(24):
    for j in range(4):
        data.append([i + 1, location[j], random.uniform(13, 35), random.randint(28079001, 28079100)])
df = pd.DataFrame(data, columns=head)
df
```

```python
sns.relplot(x='hour', y='no_2_mean', data=df,
           kind='line')
plt.show()
```

### Subgroups

```python
sns.relplot(x='hour', y='no_2_mean', data=df,
           kind='line', style='location',
           hue='location', markers=True)
plt.show()
```

### Multiple obseravations per x-value

### Все значения уровня оксида азота объединены и жирной линией показано среднее значение, область - доверительный интервал

```python
sns.relplot(x='hour', y='no_2_mean', data=df, kind='line')
plt.show()
```

### Заменим доверительный интервал на стандартное отклонение

```python
sns.relplot(x='hour', y='no_2_mean', data=df, kind='line',
           ci='sd')
plt.show()
```

## Adding titles and labels 1

### Add title to FacetGrid object

```python
sns.set_context('notebook')
g = sns.catplot(x='smoker', y='total_bill', data=tips,
               kind='box')
g.fig.suptitle('Mean total bill for smokers and no smokers',
              y=1.03)
plt.show()
```

## Adding titles and labels 2

### Adding title to AxesSubplot object

```python
g = sns.boxplot(x='smoker', y='total_bill', data=tips)
g.set_title('Mean total bill for smokers and no smokers', y=1.04)
plt.show()
```

### Adding titles to Subplots

```python
g = sns.catplot(x='smoker', y='total_bill', data=tips,
           kind='box', col='time')
g.fig.suptitle('Mean total bill for smokers and no smokers\n by day',
              y=1.05) # main title for all plots
g.set_titles('This is {col_name}')
plt.show()
```

### Adding axis labels

```python
g = sns.catplot(x='smoker', y='total_bill', data=tips,
               kind='box')
g.set(xlabel='Smokers',
     ylabel='Bill')
plt.show()
```

### Rotating x=axis tick labels

```python
g = sns.catplot(x='smoker', y='total_bill', data=tips,
                kind='box')
plt.xticks(rotation=45)
plt.show()
```
