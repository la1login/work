## Sorting and subsetting

### Single value

```python
df.sort_values('Age', ascending=False)
```

### Multiple values

```python
df.sort_values(['Age', 'Annual Income (k$)'], ascending=[True, False])
```

### Subsetting

```python
df[['Age', 'Gender']]
```

```python
df[df.Age > 40]
```

```python
df[df.Age == 30]
```

```python
df[(df.Age == 18) & (df.Gender == 'Male')]
```

```python
df[a & b]
```

```python
df[df.Age.isin(range(28, 31))]
```

```python
df[(df.Age.isin([28, 31])) & (df.Gender == 'Male')]
```
