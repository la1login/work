## Plotting time-series with different variables

```python
fig, ax = plt.subplots()
ax.plot(df.index, df['co2'], color='r')
ax.plot(df.index, df['temp'], color='b')
plt.show()
```

### Isn't normally looks

```python
fig, ax = plt.subplots()
ax.plot(df.index, df['co2'], color='r')
ax.set_xlabel('Time')
ax.set_ylabel('CO2', color='r')
ax.tick_params('y', colors='r')
ax2 = ax.twinx() # creating new ax object which similar to first
ax2.plot(df.index, df['temp'], color='b')
ax2.set_ylabel('Temp', color='b')
ax2.tick_params('y', colors='b')
plt.show()
```

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color = color)
    axes.tick_params('y', colors=color)
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, df.index, df['co2'], 'blue', "Time", 'CO2')
ax2 = ax.twinx()
plot_timeseries(ax2, df.index, df['temp'], 'red', "Time", 'Temp')
```

```python
df_6month = df['2016-06-06': '2016-12-06']
fig, ax = plt.subplots()
plot_timeseries(ax, df_6month.index, df_6month['co2'], 'blue', "Time", 'CO2')
ax2 = ax.twinx()
plot_timeseries(ax2, df_6month.index, df_6month['temp'], 'red', "Time", 'Temp')
```

## Annotating time-series data

```python
fig, ax = plt.subplots()
plot_timeseries(ax, df.index, df['co2'], 'blue', "Time", 'CO2')
ax2 = ax.twinx()
plot_timeseries(ax2, df.index, df['temp'], 'red', "Time", 'Temp')
ax2.annotate('Annotation', xy=[pd.Timestamp('2016-10-06'), 0.85])
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, df.index, df['co2'], 'blue', "Time", 'CO2')
ax2 = ax.twinx()
plot_timeseries(ax2, df.index, df['temp'], 'red', "Time", 'Temp')
ax2.annotate('Annotation', xy=[pd.Timestamp('2016-10-06'), 0.85],
            xytext=(pd.Timestamp('2016-08-06'), 0.8));
```

### Adding arrows to annotation

```python
fig, ax = plt.subplots()
plot_timeseries(ax, df.index, df['co2'], 'blue', "Time", 'CO2')
ax2 = ax.twinx()
plot_timeseries(ax2, df.index, df['temp'], 'red', "Time", 'Temp')
ax2.annotate('Annotation', xy=[pd.Timestamp('2016-10-06'), 0.9],
            xytext=(pd.Timestamp('2016-08-06'), 0.8),
            arrowprops={})
plt.show()
```

### Customizing arrow propeties

```python
fig, ax = plt.subplots()
plot_timeseries(ax, df.index, df['co2'], 'blue', "Time", 'CO2')
ax2 = ax.twinx()
plot_timeseries(ax2, df.index, df['temp'], 'red', "Time", 'Temp')
ax2.annotate('Annotation', xy=[pd.Timestamp('2016-10-06'), 0.9],
            xytext=(pd.Timestamp('2016-08-06'), 0.8),
            arrowprops={'arrowstyle':'->',
                       'color':'gray'})
plt.show()
```

# 3 Quantitaative comparisons - количественные сравнения

## Bar charts - значение переменной в различных условиях

```python
head = ['country', 'gold', 'silver', 'bronze']
data = [
    ['United States', 137, 52, 67],
    ['Germany', 47, 43, 67],
    ['Great Britain', 64, 55, 26],
    ['Russia', 50, 28, 35],
    ['China', 44, 30, 35],
    ['France', 20, 55, 21],
    ['Australia', 23, 34, 25],
    ['Italy', 8, 38, 24],
    ['Canada', 4, 4, 61],
    ['Japan', 17, 13, 34],
]
df = pd.DataFrame(data, columns=head)
df = df.set_index('country')
df
```

```python
fig, ax = plt.subplots()
ax.bar(df.index, df['gold'])
plt.show()
```

### Rotate ticks

```python
fig, ax = plt.subplots()
ax.bar(df.index, df['gold'])
ax.set_xticklabels(df.index, rotation=90)
ax.set_ylabel('Number of gold medals')
plt.show()
```

### Different columns с накоплением

```python
fig, ax = plt.subplots()
ax.bar(df.index, df['gold'])
ax.bar(df.index, df['silver'], bottom=df['gold'])
ax.set_xticklabels(df.index, rotation=90)
ax.set_ylabel('Number of gold and silver medals')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(df.index, df['gold'])
ax.bar(df.index, df['silver'], bottom=df['gold'])
ax.bar(df.index, df['bronze'], bottom=df['gold'] + df['silver'])
ax.set_xticklabels(df.index, rotation=90)
ax.set_ylabel('Number of all medals')
plt.show()
```

### Adding legend

```python
fig, ax = plt.subplots()
ax.bar(df.index, df['gold'], label='gold')
ax.bar(df.index, df['silver'], bottom=df['gold'], label='silver')
ax.bar(df.index, df['bronze'], bottom=df['gold'] + df['silver'],
      label='bronze')
ax.set_xticklabels(df.index, rotation=90)
ax.set_ylabel('Number of all medals')
ax.legend()
plt.show()
```

## Histograms - распределение значений внутри переменной

```python
df = pd.read_csv('customers.csv')
df
```

```python
df_inc = df[['Annual Income (k$)']]
```

```python
df_spd = df[['Spending Score (1-100)']]
```

Different between income and spending

```python
fig, ax = plt.subplots()
ax.bar('Income', df_inc['Annual Income (k$)'].mean())
ax.bar('Spend', df_spd['Spending Score (1-100)'].mean())
ax.set_ylabel('Money')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df_inc['Annual Income (k$)'], label='income', bins=5)
ax.hist(df_spd['Spending Score (1-100)'], alpha=0.7, label='spend',
       bins=5)
ax.set_xlabel('Money and scores')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df_inc['Annual Income (k$)'], label='income', 
        bins=[20, 40, 60, 80, 100])
ax.hist(df_spd['Spending Score (1-100)'], alpha=0.7, label='spend',
       bins=range(20, 120, 20))
ax.set_xlabel('Money and scores')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df_inc['Annual Income (k$)'], label='income', 
        bins=[20, 40, 60, 80, 100],
       histtype='step')
ax.hist(df_spd['Spending Score (1-100)'], alpha=0.7, label='spend',
       bins=range(20, 120, 20),
       histtype='step')
ax.set_xlabel('Money and scores')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

## Automating figures from data

### Допустим, мы не знаем, сколько конкретно значений у нас есть. Для этого создаем объект Series из интересующего столбца и вызываем метод .unique( )
### Далее в цикле проходим по данному объекту и выполняем нужные действия

```python
df = pd.read_csv('customers.csv')
unique = df['Age'].unique()
unique
```

```python
df['Age_range10'] = df['Age'] % 7
```

```python
unique = df['Age_range10'].unique()
unique
```

```python
fig, ax = plt.subplots()
for age in unique:
    sub_df = df[df['Age_range10'] == age]
    ax.bar(age, sub_df['Annual Income (k$)'].mean(), 
           yerr=sub_df['Annual Income (k$)'].std())
ax.set_ylabel('Annual Income')
ax.set_xticklabels(unique)
plt.show()
```
