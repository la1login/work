# Plotting data with Matplotlib

```python
plt.plot(df['Annual Income (k$)'], df.Age,
         label='Income vs Age')
plt.plot(df['Annual Income (k$)'], df['Spending Score (1-100)'], 
         alpha=0.7, label='Income vs Spending')
plt.xlabel('Income')
plt.ylabel('Age or Spending')
plt.title('Income vs Age and Spending', fontsize=20, color='green')
plt.legend()
plt.text(20, 20, 'Annotation')
plt.show()
```

##  Styling graphs

```python
plt.plot(df.CustomerID, df.Age, color='violet', linewidth=3)
plt.show()
```

```python
plt.plot(df.CustomerID, df.Age, color='violet', linewidth=2,
        linestyle='--')
plt.show()
```

```python
plt.plot(df.CustomerID, df.Age, color='violet',
        marker='*')
plt.show()
```

### Setting style
Set for all next graphics until style will change

```python
plt.style.use('ggplot')
plt.plot(df.CustomerID, df.Age, color='violet')
plt.show()
```

```python
plt.style.use('default')
plt.plot(df.CustomerID, df.Age, color='violet')
plt.show()
```

# Other types of graphics

## Scatter plot

```python
plt.scatter(df.Age, df['Annual Income (k$)'])
plt.show()
```

### Kwargs

```python
plt.scatter(df.Age, df['Annual Income (k$)'],
           color='violet', marker='s', alpha=0.5)
plt.show()
```

## Bar plot

```python
plt.bar(df.Gender, df['Annual Income (k$)'])
plt.show()
```

```python
plt.barh(df.Gender, df['Annual Income (k$)'])
plt.show()
```

### Adding error bars

```python
plt.bar(df.Gender, df.Age,
           yerr=df.Age.std())
plt.show()
```

### Stacked bar charts

```python
plt.bar(df.Gender, df['Annual Income (k$)'])
plt.bar(df.Gender, df['Spending Score (1-100)'], 
        bottom=df['Annual Income (k$)'], alpha=0.5)
plt.show()
```

## Histograms

```python
plt.hist(df['Annual Income (k$)'], bins=15)
plt.show()
```

 ### Changing range of values

```python
plt.hist(df['Annual Income (k$)'], 
         range=[20, 100], bins=15)
plt.show()
```

### Normalizing

```python
plt.hist(df['Annual Income (k$)'], density=True)
plt.hist(df['Spending Score (1-100)'], density=True, alpha=0.5)
plt.show()
```
