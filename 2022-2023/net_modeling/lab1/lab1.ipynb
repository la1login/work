2. Сгенерировать 1000 случайных чисел с любым известным распределением и построить их гистограмму, математическое ожидание, дисперсию.

Генерируем выборку размера 1000 из нормального распределения с МО = 1 и СКО = 5
А также генерируем вспомогательный np-массив для отрисовки кривой нормального распределения

```python
sample = stats.norm.rvs(1, 5, 1000)
x = np.arange(-15, 16, 0.001)
```

Для коррекции значений по оси Y на ум первым пришло решение через объектный подход и получаем две разных оси Y на одной оси Х.

```python
plt.style.use('default')
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.hist(sample, bins=20)
ax2.plot(x, stats.norm.pdf(x, 1, 5), c='red', linewidth=2)
ax1.set_ylabel('Выборка')
ax2.set_ylabel('Плотность нормального распределения', c='red')
plt.show()
```

```python
print('Математическое ожидание: ', sample.mean(),
      '\nДисперсия: ', sample.var())
```

3. Сгенерировать случайную точку, равномерно распределенную в квадрате со стороной а.

Будем работать с квадратом со стороной a = 5, генерируем точку в нем

```python
x = np.random.uniform(1, 6, 1)[0]
y = np.random.uniform(1, 6, 1)[0]
print(x, y)
```

Отрисовываем квадрат и точку

```python
plt.style.use('bmh')
plt.plot([1, 6], [1, 1], 
         [1, 1], [1, 6], 
         [1, 6], [6, 6], 
         [6, 6], [6, 1],
        color='r', linewidth=3)
plt.scatter(x, y, color='g', linewidth=5)
plt.title('Квадрат со стороной a = 5')
plt.show()
```

И функция для любого значения а

```python
def square_point(a):
    if a > 0:
        a += 1
        x, y = np.random.uniform(1, a, 1)[0], np.random.uniform(1, a, 1)[0]
        print('Координаты точки: ', round(x, 4), round(y, 4))
        plt.plot([1, a], [1, 1], 
                 [1, 1], [1, a], 
                 [1, a], [a, a], 
                 [a, a], [a, 1],
                color='r', linewidth=3)
        plt.scatter(x, y, color='g', linewidth=5)
        plt.title('Квадрат со стороной a = ' + str(a - 1))
        plt.show()
    else:
        print('Неверная длина стороны квадрата')
```

```python
square_point(1)
```
