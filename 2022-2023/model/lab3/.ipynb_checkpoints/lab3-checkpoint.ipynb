### Применение распределений для преобразования случайных величин

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math
```

Для начала просто нарисуем круг

```python
theta = np.linspace(0, 2 * np.pi, 150)
 
radius = 3
 
a = radius * np.cos(theta)
b = radius * np.sin(theta)
 
figure, axes = plt.subplots(1)
 
axes.plot(a, b)
axes.set_aspect(1)
plt.show()
```

### Способ 1
Генерируем выборку по случайному радиусу и углу.

```python
r = stats.uniform.rvs(0, 3, size=1000)
angle = stats.uniform.rvs(0, 360, size=1000)
```

Переводим координаты в полярные

```python
x = [r[i] * math.cos(angle[i]) for i in range(1000)]
y = [r[i] * math.sin(angle[i]) for i in range(1000)]
```

Рисуем круг и точки на графике

```python
theta = np.linspace(0 , 2 * np.pi, 150)
 
radius = 3
 
a = radius * np.cos(theta)
b = radius * np.sin(theta)
 
figure, axes = plt.subplots( 1 )
 
axes.plot(a, b)
axes.set_aspect(1)
plt.scatter(x, y)
plt.title('Способ №1')
plt.show()
```

Все точки лежат строго в пределах окружности

### Способ 2
Генерируем значения в пределах квадрата

```python
x1 = stats.uniform.rvs(-3, 6, 1000)
y1 = stats.uniform.rvs(-3, 6, 1000)
```

Переводим координаты в полярные и рисуем график

```python
theta = np.linspace(0, 2 * np.pi , 150)
 
radius = 3
 
a = radius * np.cos(theta)
b = radius * np.sin(theta)
 
figure, axes = plt.subplots(1)
 
axes.plot( a, b, linewidth=3, c='r')
plt.plot(-3, -3, -3, 3)
axes.set_aspect( 1 )
plt.plot([-3, -3], [-3, 3],
         [-3, 3], [3, 3],
         [3, 3], [3, -3],
         [-3, 3], [-3, -3],
         c='r',
         linewidth=3)
plt.scatter(x1, y1)
plt.title( 'Способ №2. Все точки' )
plt.show()
```

Функция для удаления лишних точек

```python
for_del = []
for i in range(len(x)):
    if (x1[i] ** 2 + y1[i] ** 2) > 9:
        for_del.append(i)
```

```python
x_clear = np.delete(x1, for_del)
y_clear = np.delete(y1, for_del)
```

Отрисовка после удаления

```python
theta = np.linspace(0, 2 * np.pi, 150)
 
radius = 3
 
a = radius * np.cos(theta)
b = radius * np.sin(theta)
 
figure, axes = plt.subplots(1)
 
axes.plot(a, b,
          linewidth=3,
          c='r')
plt.plot(-3, -3, -3, 3)
axes.set_aspect( 1 )
plt.plot([-3, -3], [-3, 3],
         [-3, 3], [3, 3],
         [3, 3], [3, -3],
         [-3, 3], [-3, -3], 
         color='r',
         linewidth=3)
plt.scatter(x_clear, y_clear)
plt.title( 'Способ №2. Удаление лишних точек' )
plt.show()
```

```python
print('Способ 1: \nВыборочное среднее по х =', np.mean(x),  'дисперсия по x =', np.var(x))
print('\nВыборочное среднее по y =', np.mean(y),  'дисперсия по y =', np.var(y))
print('\nСпособ 2: \nВыборочное среднее по х =', np.mean(x_clear),  'дисперсия по x =', np.var(x_clear))
print('\nВыборочное среднее по y =', np.mean(y_clear),  'дисперсия по y =', np.var(y_clear))
```

Координаты точки, до которой ищем расстояние

```python
dot_x = 20
dot_y = 0
```

```python
theta = np.linspace(0 , 2 * np.pi, 150)
 
radius = 3
 
a = radius * np.cos(theta)
b = radius * np.sin(theta)
plt.style.use('default')
figure, axes = plt.subplots(1) 
axes.plot(a, b)
plt.plot(-3, -3, -3, 3)
axes.set_aspect( 1 )
plt.plot([-3, -3], [-3, 3],
         [-3, 3], [3, 3],
         [3, 3], [3, -3],
         [-3, 3], [-3, -3], color='r')
plt.scatter(x_clear, y_clear)
plt.scatter(dot_x, dot_y)
plt.show()
```

Найдем расстояние от точки до точек из первого способа. Построим гистограмму распределения

```python
distances1 = [math.sqrt((x[i] - dot_x)**2 + (y[i] - dot_y)**2) for i in range(len(x))]
plt.hist(distances1, bins=20)
plt.show()
```

То же самое для второго способа генерирования точек

```python
distances2 = [math.sqrt((x_clear[i] - dot_x)**2 + (y_clear[i] - dot_y)**2) for i in range(len(x_clear))]
plt.hist(distances2, bins=20)
plt.show()
```

Функция для поиска расстояния между случайными точками

```python
def find_random_dist(x, y, size):
    inner_dist = []
    for i in range(size):
        num1 = round(np.random.uniform(0, size - 1, 1)[0])
        num2 = round(np.random.uniform(0, size - 1, 1)[0])
        inner_dist.append(math.sqrt((x[num1] - x[num2])**2 + (y[num1] - y[num2])**2))
    return inner_dist
```

```python
innder_dist1 = find_random_dist(x, y, 1000)
```

Строим гистограмму плотности для первого способа

```python
plt.hist(inner_dist1, bins=15)
plt.show()
```

Повторяем для второго способа

```python
innder_dist2 = find_random_dist(x_clear, y_clear, len(x_clear))
```

```python
plt.hist(inner_dist2, bins=10)
plt.show()
```
