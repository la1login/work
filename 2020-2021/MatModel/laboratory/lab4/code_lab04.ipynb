```python
import numpy
from scipy. integrate import odeint
import matplotlib.pyplot as pl
import math

tmax = 37
step = 0.05
y0 = [1, 1.2]

w2 = 3.5

t = numpy.arange(0, tmax, step)

def W(y, t):
    y1,y2=y
    return[y2, -w2*y1]

w1 = odeint(W, y0, t)
y11=w1[:,0]
y21=w1[:,1]

fig1 = pl.figure(facecolor = 'white')
pl.plot(y11, y21, linewidth = 2)
pl.ylabel("x'")
pl.xlabel("x")
pl.grid(True)
pl.show()
```

```python
w2 = 3
g = 7

def W(y,t):
    y1,y2=y
    return[y2, -w2*y1 - g*y2]
t = numpy.arange(0, tmax, step)

w1 = odeint(W, y0, t)
y11 = w1[:,0]
y21 = w1[:,1]

fig2 = pl.figure(facecolor='white')
pl.plot(y11,y21, linewidth = 2)
pl.ylabel("x'")
pl.xlabel("x")
pl.grid(True)
pl.show()
```

```python
w2 = 2

g = 5

def f(t):
    f = 2*math.sin(6*t)
    return f

def W(y,t):
    y1,y2 = y
    return[y2, f(t)-w2*y1-g*y2]

t = numpy.arange(0, tmax, step)

w1 = odeint(W, y0, t)

y11 = w1[:,0]
y21 = w1[:,1]

fig3 = pl.figure(facecolor = 'white')

pl.plot(y11, y21, linewidth = 2)
pl.ylabel("x'")
pl.xlabel("x")
pl.grid(True)
pl.show()
```
